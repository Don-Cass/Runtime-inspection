```python
import pandas as pd
import matplotlib.pyplot as plt

# CSV 파일 로드
data_array = pd.read_csv('execution_times_array.csv')
data_heap = pd.read_csv('execution_times_heap.csv')

# 각 메소드 별 시각화
plt.figure(figsize=(12, 6))

# 배열 삽입 시간 시각화

# 나머지 메소드에 대한 그래프도 비슷한 방식으로 작성
plt.subplot(2, 3, 1)
plt.scatter(data_array['Size'], data_array['Time_insert (nanoseconds)'], label='Array Insert Time')
plt.xlabel('Size')
plt.ylabel('Time (nanoseconds)')
plt.title('Array Insert Time vs Size')

plt.tight_layout()
plt.show()


```

```python
plt.figure(figsize=(15, 5))

# 배열과 힙 삽입 시간 비교
plt.subplot(1, 3, 1)
plt.scatter(data_array['Size'], data_array['Time_insert (nanoseconds)'], alpha=0.1, label='Array Insert Time', s=10)
plt.scatter(data_heap['Size'], data_heap['Time_insert (nanoseconds)'], alpha=0.1, label='Heap Insert Time', s=10)
plt.xlabel('Size')
plt.ylabel('Time (nanoseconds)')
plt.title('Insert Time Comparison')
plt.legend()

# 배열과 힙 최대값 출력 시간 비교
plt.subplot(1, 3, 2)
plt.scatter(data_array['Size'], data_array['Time_getMax (nanoseconds)'], alpha=0.1, label='Array GetMax Time', s=10)
plt.scatter(data_heap['Size'], data_heap['Time_getMax (nanoseconds)'], alpha=0.1, label='Heap GetMax Time', s=10)
plt.xlabel('Size')
plt.ylabel('Time (nanoseconds)')
plt.title('GetMax Time Comparison')
plt.legend()

# 배열과 힙 최대값 삭제 시간 비교
plt.subplot(1, 3, 3)
plt.scatter(data_array['Size'], data_array['Time_deleteMax (nanoseconds) '], alpha=0.1, label='Array DeleteMax Time', s=10)
plt.scatter(data_heap['Size'], data_heap['Time_deleteMax (nanoseconds) '], alpha=0.1, label='Heap DeleteMax Time', s=10)
plt.xlabel('Size')
plt.ylabel('Time (nanoseconds)')
plt.title('DeleteMax Time Comparison')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
def calculate_average_times(data):
    grouped = data.groupby('Size')
    return grouped.mean()

# 배열과 힙 데이터에 대해 평균 시간 계산
average_times_array = calculate_average_times(data_array)
average_times_heap = calculate_average_times(data_heap)

# 결과 출력
print("Array Average Times:\n", average_times_array)
print("Heap Average Times:\n", average_times_heap)
```

```python
# 특정 크기 (예: 5000000)에서 각 메소드의 평균 실행 시간 계산
specific_size = 5000000

# 배열과 힙 데이터에서 특정 크기에 해당하는 데이터만 필터링
specific_data_array = data_array[data_array['Size'] == specific_size]
specific_data_heap = data_heap[data_heap['Size'] == specific_size]

# 평균 실행 시간 계산
average_times_array_specific = specific_data_array.mean()
average_times_heap_specific = specific_data_heap.mean()

average_times_array_specific, average_times_heap_specific
```

```python
# 5,000,000에 가장 가까운 크기 찾기
closest_size_array = data_array.iloc[(data_array['Size'] - specific_size).abs().argsort()[:1]]['Size'].iloc[0]
closest_size_heap = data_heap.iloc[(data_heap['Size'] - specific_size).abs().argsort()[:1]]['Size'].iloc[0]

# 가장 가까운 크기에 해당하는 데이터 필터링
closest_data_array = data_array[data_array['Size'] == closest_size_array]
closest_data_heap = data_heap[data_heap['Size'] == closest_size_heap]

# 평균 실행 시간 계산
average_times_array_closest = closest_data_array.mean()
average_times_heap_closest = closest_data_heap.mean()

average_times_array_closest, average_times_heap_closest
```
